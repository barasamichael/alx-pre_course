```python
# Import required libraries
import geopandas as gpd
import pandas as pd
import warnings
import matplotlib.pyplot as plt

# Suppress warnings
warnings.filterwarnings('ignore')
```

```python
# Rename the column header to match the one in our csv file
old_column_name = 'county'
new_column_name = 'residence_county'
gdf = gdf.rename(columns={old_column_name: new_column_name})

# Display modified geopandas dataframe
gdf.head()
```

```python
# Merge the shapefile and CSV data
merged_df = gdf.merge(data, left_on='residence_county', right_on='residence_county')
```

```python
# Set utility functions
def set_ticks(ax):
    "Accepts an Axes object and removes backticks"
    ax.set_xticks([])
    ax.set_yticks([])

def add_compass(ax):
    "Accepts an Axes object and adds compass directions"
    ax.annotate('N', xy=(0.5, 0.95), xycoords='axes fraction',
            fontsize=12, ha='center', va='center')
    ax.annotate('S', xy=(0.5, 0.05), xycoords='axes fraction',
            fontsize=12, ha='center', va='center', rotation=180)
    ax.annotate('E', xy=(0.95, 0.5), xycoords='axes fraction',
            fontsize=12, ha='center', va='center', rotation=-90)
    ax.annotate('W', xy=(0.05, 0.5), xycoords='axes fraction',
            fontsize=12, ha='center', va='center', rotation=90)

def get_plot(attribute, theme = 'YlOrRd'):
    """Returns a plot of the provided attribute"""
    fig, ax = plt.subplots(figsize=(10, 10))
    merged_df.plot(column = attribute, cmap = theme, linewidth = 0.8, ax = ax, 
                   edgecolor = '0.8', legend=True)
    
    # Assign labels to counties 
    gdf['centroid'] = gdf.geometry.centroid
    for x, y, label in zip(gdf.centroid.x, gdf.centroid.y, gdf['residence_county']):
        ax.text(x, y, label, fontsize=8, ha='center', va='center')

    set_ticks(ax)
    add_compass(ax)
    
    # set plot title
    ax.set_title(f'Graduated Map - {attribute}', 
                 fontdict={'fontsize': '16', 'fontweight' : '3', 'color' : 'green'})
    
    return plt
```

```python
plot = get_plot('Severity of Poverty (%)', 'Accent')
plot.show()
```

```python
plot = get_plot('Population (ths)', 'CMRmap')
plot.show()
```

```python
plot = get_plot('Number of Poor (ths)', 'OrRd')
plot.show()
```
