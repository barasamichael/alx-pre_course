## <span style = "color:#1A237E;">Hypothesis One</span>
### <span style = "color:green;">Null Hypothesis</span>
Households in counties with lower headcount rates are less likely to access credit.
### <span style = "color:green;">Alternative Hypothesis</span>
Households in counties with higher headcount rates are less likely to access credit.
### <span style = "color:green;">Relevance</span>
In Kenya, poverty is more prevalent in some counties compared to others. 
Therefore, it is hypothesized that households in poorer counties will have less access to credit, 
which may worsen their poverty status.

```python
# import necessary libraries
import pandas as pd
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
# Load data into a pandas dataframe
data = pd.read_csv('overall_poverty_est.csv')
data.head()
```

```python
# Display the columns and rows
data.shape
```

```python
# Display the column headers of dataset
data.columns
```

```python
# Display statistical summary of dataset
data.describe()
```

### <span style = "color:green;">Headcount Rate vs Proportion of households that sought credit</span>

```python
# Scatter plot of Headcount Rate vs Proportion of households that sought credit
plt.scatter(data['Headcount Rate (%)'], data['Proportion of households that sought credit (%)'])
plt.xlabel('Headcount Rate (%)')
plt.ylabel('Proportion of households that sought credit (%)')
plt.show()
```

### <span style = "color:green;">Correlation Coefficient and P-value</span>

```python
# Calculate the correlation coefficient and p-value
r, p = stats.pearsonr(data['Headcount Rate (%)'], data['Proportion of households that sought credit (%)'])
print('Correlation coefficient:', r)
print('p-value:', p)
```

#### <span style = "color:brown;">Observations and Inferences</span>
The correlation coefficient is `-0.4777`, which indicates _a moderate negative correlation between 
the `Headcount Rate (%)` and the `Proportion of households that sought credit (%)`_. The `p-value` is 
`0.0379`, which is _below_ the commonly used threshold of `0.05`. This suggests that _the observed 
correlation between the two variables is statistically significant_, and not likely to have 
occurred by chance.

### <span style = "color:green;">Mean Proportions</span>

```python
# Calculate the mean proportion of households that sought and accessed credit for counties with Headcount Rate >= 50%
high_hr_mean = data[data['Headcount Rate (%)'] >= 50]['Proportion of households that sought and accessed credit (%)'].mean()
print('Mean proportion of households that sought and accessed credit for counties with Headcount Rate >= 50%:', high_hr_mean)
```

```python
# Calculate the mean proportion of households that sought and accessed credit for counties with Headcount Rate < 50%
low_hr_mean = data[data['Headcount Rate (%)'] < 50]['Proportion of households that sought and accessed credit (%)'].mean()
print('Mean proportion of households that sought and accessed credit for counties with Headcount Rate < 50%:', low_hr_mean)
```

#### <span style = "color:brown;">Observations and Inferences</span>
The `mean proportion of households that sought and accessed credit` for counties with `Headcount 
Rate >= 50%` is `72.17%`, while the `mean proportion of households that sought and accessed credit` 
for counties with `Headcount Rate < 50%` is `89.50%`.

### <span style = "color:green;">Two-Sample T-test</span>

```python
# Perform a two-sample t-test to determine if the difference in means is statistically significant
t_value, p_value = stats.ttest_ind(data[data['Headcount Rate (%)'] >= 50]['Proportion of households that sought and accessed credit (%)'],
                       data[data['Headcount Rate (%)'] < 50]['Proportion of households that sought and accessed credit (%)'],
                       equal_var=False)
print('t-value:', t_value)
print('p-value:', p_value)
```

#### <span style = "color:brown;">Observations and Inferences</span>
The `t-value` is `-2.368`, which indicates that _the means of the two groups are significantly 
different from each other_. The `p-value` is `0.038`, which is `below the commonly used threshold` of `0.05`. 
This suggests that _the difference in means between the two groups is statistically significant_, and not 
likely to have occurred by chance.
### <span style = "color:green;">Conclusion</span>
Therefore, we can conclude that _the proportion of households that sought and accessed credit `is 
significantly different`- for counties with `Headcount Rate >= 50%` compared to those with `Headcount Rate < 50%`.
