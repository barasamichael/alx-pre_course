## <span style = "color:#1A237E;">Hypothesis Five</span>
### <span style = "color:green;">Null Hypothesis</span>
The distribution of the poor is not correlated with credit access.
### <span style = "color:green;">Alternative Hypothesis</span>
The distribution of the poor is correlated with credit access.
### <span style = "color:green;">Relevance</span>
Poverty is distributed unevenly across different parts of Kenya. This hypothesis seeks to explore whether there is a relationship between the distribution of poverty and credit access in Kenya.

```python
# Import required libraries
import pandas as pd
from scipy.stats import pearsonr
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the data
data = pd.read_csv('overall_poverty_est.csv')
data.head()
```

```python
# Display statistical summary
data.describe()
```

```python
# Display number of  rows and columns
data.shape
```

```python
# display column headers
data.columns
```

```python
# visualize correlation between variables
sns.pairplot(data=data, vars=["Distribution of the Poor (%)", "Proportion of households that sought and accessed credit (%)"],
             height=1.5, aspect=1.5)
plt.show()
```

```python
# visualize distribution of the poor
sns.histplot(data=data, x="Distribution of the Poor (%)")
plt.show()
```

```python

# visualize credit access
sns.histplot(data=data, x="Proportion of households that sought and accessed credit (%)")
plt.show()
```

```python
# Calculate the correlation coefficient and p-value
corr, p_value = pearsonr(
    data['Proportion of households that sought and accessed credit (%)'], 
    data['Distribution of the Poor (%)']
)
```

```python
# Print the results
print('Correlation coefficient:', corr)
print('p-value:', p_value)
```

### <span style = "color:green;">Observations and Inferences</span>

* A correlation coefficient of `0.005` indicates _a very weak positive correlation_ between the two 
variables, **while** the p-value of `0.972` suggests that _the correlation is not statistically significant_ 
at the standard significance level of `0.05`. 
* This means that there is no evidence to suggest that the two variables are related to each other.
